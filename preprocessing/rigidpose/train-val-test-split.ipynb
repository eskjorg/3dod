```python
with open(os.path.join(SIXD_PATH, 'models', 'models_info.yml'), 'r') as f:
    models_info = yaml.load(f, Loader=yaml.CLoader)
```

```python
nbr_seqs = 92
df_raw = OrderedDict([(obj, [False]*nbr_seqs) for obj in sorted([models_info[obj_idx]['readable_label'] for obj_idx in models_info.keys()])])
# for seq in range(19,20):
for seq in range(nbr_seqs):
    for subset in ['train', 'val', 'test']:
        seq_path = os.path.join(SIXD_PATH, subset, '{:04d}'.format(seq))
        if os.path.exists(seq_path):
            break
    else:
        assert False
    with open(os.path.join(seq_path, 'global_info.yml'), 'r') as f:
        global_info_yaml = yaml.load(f, Loader=yaml.CLoader)
    for obj in global_info_yaml['obj_annotated_and_present']:
        df_raw[obj][seq] = True
```

```python
df = DataFrame(df_raw)
```
